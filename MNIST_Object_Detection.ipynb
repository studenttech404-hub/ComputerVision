```python
!pip install ultralytics
from ultralytics import YOLO

# Load a base YOLOv8 model
model = YOLO("yolov8n.pt")  # options: yolov8n.pt, yolov8s.pt, yolov8m.pt, etc.

# Train the model
model.train(data=f"{dataset.location}/data.yaml", epochs=30, imgsz=640, batch=16)
```

```python
# Run validation
model.val()
```

```python
# Predict on a sample image
results = model.predict(source="https://ultralytics.com/images/bus.jpg", conf=0.5)

# Show results
results[0].show()
```
